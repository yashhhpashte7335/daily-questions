```python
import matplotlib.pyplot as plt

```

```python
xAxis = list(range(1,20))
yAxis = [22,34,55,65,44,33,22,4,44,55,66,45,22,33,87,76,65,44,97]
plt.plot(xAxis,yAxis,marker='o',linewidth=2,ls=':',color='green',markerfacecolor='red')
plt.plot(xAxis,yAxis,label=["count of days","price in $"])
plt.xlabel("count of days")
plt.ylabel("price in ")

plt.title("product price")
```

```python
cities=['nashik','mumbai','thane','junnar','kalwa']
rainfall=[300,44,130,100,250]
plt.bar(cities,rainfall,color='skyblue',edgecolor="black")

plt.xlabel("")
plt.ylabel("Average monthly rainfall (mm)")
plt.title("Average monthly rainfall in different cities")
plt.show()
```
